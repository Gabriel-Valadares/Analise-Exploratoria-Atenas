Verificando a tabela dos resumos estat√≠sticos, suspeita-se da presen√ßa de 
outlier nos dados, haja vista os valores m√°ximos para pre√ßo e tempo m√≠nimo de estadia serem muito elevados em rela√ß√£o a 75% e 50%. Ou seja:

* A vari√°vel `Price` possui 75% de seus valores abaixo de 100, mas seu valor m√°ximo √© 8000.
* A vari√°vel `minimum_nights` possui 75% de seus valores abaixo de 2, mas seu valor m√°ximo √© 1000, o que √© acima de 365 dias no ano.

Os valores que proporei como limite de corte na tabela ser√£o determinados conforme fa√ßa sentido com o gr√°fico de box plot (como a an√°lise est√° um n√≠vel superficial, n√£o ser√° empregado um m√©todo cient√≠fico formal).

* *`price`*: Valores acima de 3 vezes em rela√ß√£o ao valor m√°ximo de 75% das entradas nessa coluna.
* *`minimum_nights`*: Valores acima de 2 semanas de estadia.

```
# minimum_nights
df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

# ver quantidade de valores acima de 15 dias para minimum_nights
print("\nminimum_nights: valores acima de 15:")
print("{} entradas".format(len(df[df.minimum_nights > 15])))
print("{:.4f}%".format((len(df[df.minimum_nights > 15]) / df.shape[0])*100))
```

```
# Price
df.price.plot(kind='box', vert=False, figsize=(15, 3),)
plt.show()

# ver quantidade de valores acima de 300 para price
print("\nprice: valores acima de 300")
print("{} entradas".format(len(df[df.price > 300])))
print("{:.4f}%".format((len(df[df.price > 300]) / df.shape[0])*100))

# df.price.plot(kind='box', vert=False, xlim=(0,1300), figsize=(15,3));
```



---


Vamos eliminar os outliers e a coluna neighbourhood_group, que est√° vazia.

```
# Eiminando outliers em um novo Dataframe
df_clean = df.copy()

# Baseado no Price
df_clean.drop(df_clean[df_clean.price > 300].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 15].index, axis=0, inplace=True)

# Eliminando o neighbourhood_group, pois √© uma coluna vazia
df_clean.drop('neighbourhood_group', axis=1, inplace=True)
```

```
# Novos histogramas sem os outliers
df_clean.hist(bins=15, figsize=(15,10));
```



---


Agora a distribui√ß√£o das vari√°veis parece fazer mais sentido. Agora vamos tentar entender se h√° alguma correla√ß√£o entre as vari√°veis

### **Q5. Qual √© a correla√ß√£o existente entre as vari√°veis?**

```
# Matriz de correla√ß√£o
df_corr = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
df_corr
```

```
# Heatmap a partir das correla√ß√µes
sns.heatmap(df_corr, cmap='RdBu', square=True, linecolor='white', annot=True);
```



---

N√£o h√° rela√ß√£o minimamente forte entre as vari√°veis, a n√£o ser, por √≥bvio, a da quantidade total de avalia√ß√µes com a m√©dia mensal.

### **Q6. Qual √© o tipo de im√≥vel mais alugado no Airbnb?**

```
# Quantidade de cada tipo de im√≥vel dispon√≠vel
df_room = df_clean.groupby(['room_type'])['room_type'].count().sort_values(ascending=False)
df_room
```

```
# Porcentagem de cada tipo de im√≥vel dispon√≠vel
df_room / df.shape[0]
```



---

Ou seja, a maioria das loca√ß√µes s√£o feitas pela casa inteira, e n√£o apenas quartos privados ou compartilhados.



---
